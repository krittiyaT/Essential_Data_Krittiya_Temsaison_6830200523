## 1. ทำไมต้องมี OOP?

ลองดูปัญหานี้: เก็บข้อมูลนักศึกษาหลายคน แต่ละคนมีชื่อ, อายุ, และ GPA — ใช้ dict ที่เรียนมาแล้วเก็บได้ไหม?


````python
# ใช้ dict เก็บข้อมูลนักศึกษา (วิธีที่เรียนมาจาก Collections)
student1 = {"name": "มานี", "age": 20, "gpa": 3.75}
student2 = {"name": "สมชาย", "age": 21, "gpa": 3.20}

def print_student_info(student):
    print(f"{student['name']} อายุ {student['age']} ปี GPA {student['gpa']}")

def is_honor_student(student):
    return student['gpa'] >= 3.5

print_student_info(student1)
print("เกียรตินิยมไหม:", is_honor_student(student2))
````

วิธีนี้ **ใช้งานได้** แต่มีจุดสังเกต: ข้อมูล (`student1`) กับฟังก์ชันที่ทำงานกับข้อมูลนั้น (`print_student_info`, `is_honor_student`) ถูกเขียน**แยกกันคนละที่** ทำให้:
- ต้องจำว่ามีฟังก์ชันอะไรบ้างที่ใช้กับ dict แบบนี้ได้
- ถ้ามีข้อมูลหลายแบบ (นักศึกษา, อาจารย์, วิชา) จะมีฟังก์ชันกระจัดกระจายเต็มไปหมด
- เผลอพิมพ์ key ผิด (เช่น `student['nmae']`) จะรู้ตัวก็ตอน error เท่านั้น

**OOP (Object-Oriented Programming)** คือแนวคิดการเขียนโปรแกรมที่ **"รวมข้อมูล" และ "ฟังก์ชันที่ทำงานกับข้อมูลนั้น" ไว้ด้วยกัน** เป็นหน่วยเดียว เรียกว่า **object** — ทำให้โค้ดเป็นระเบียบขึ้นมาก เมื่อข้อมูลซับซ้อนขึ้น

## 2. Class และ Object

| | Class | Object |
|---|---|---|
| **คืออะไร** | "พิมพ์เขียว" หรือ "แบบแม่พิมพ์" ที่นิยามว่า object แบบนี้มีอะไรบ้าง | "ของจริง" ที่สร้างขึ้นมาจาก class นั้น |
| **เปรียบเทียบ** | แบบแปลนบ้าน | บ้านจริงที่สร้างจากแบบแปลนนั้น |
| **มีกี่อัน** | มีแบบเดียว (นิยามครั้งเดียว) | สร้างได้หลาย object จาก class เดียวกัน |

### โครงสร้างพื้นฐานของ class

```python
class ชื่อClass:
    def __init__(self, parameter1, parameter2):
        self.attribute1 = parameter1
        self.attribute2 = parameter2
```

**กฎการตั้งชื่อ:** ชื่อ class นิยมขึ้นต้นด้วยตัวพิมพ์ใหญ่ และไม่มี `_` คั่นคำ (เรียกว่า PascalCase) เช่น `Student`, `StudentRecord` — ต่างจากชื่อตัวแปร/ฟังก์ชันที่ใช้ snake_case


````python
class Student:
    def __init__(self, name, age, gpa):
        self.name = name    # self.name คือ attribute ของ object นี้
        self.age = age
        self.gpa = gpa

# สร้าง object จาก class Student (เรียกว่า "instantiate")
student1 = Student("มานี", 20, 3.75)
student2 = Student("สมชาย", 21, 3.20)

# เข้าถึง attribute ของแต่ละ object ด้วย .
print(student1.name, student1.age, student1.gpa)
print(student2.name, student2.age, student2.gpa)

print(type(student1))   # <class '__main__.Student'>
````

**สังเกตสิ่งสำคัญ:** `student1` และ `student2` สร้างจาก class `Student` เดียวกัน แต่เป็น **object คนละตัว** มีข้อมูลภายในของตัวเอง แก้ไข `student1` จะไม่กระทบ `student2` เลย — เหมือนบ้าน 2 หลังที่สร้างจากแบบแปลนเดียวกัน แต่เป็นบ้านคนละหลัง


````python
class Student:
    def __init__(self, name, age, gpa):
        self.name = name
        self.age = age
        self.gpa = gpa

student1 = Student("มานี", 20, 3.75)
student2 = Student("สมชาย", 21, 3.20)

student1.gpa = 4.00   # แก้ไขแค่ student1
print(f"{student1.name}: {student1.gpa}")
print(f"{student2.name}: {student2.gpa}")   # ไม่ถูกกระทบ ยังเป็น 3.20
````

### 2.1 `__init__` คืออะไร?

`__init__` เป็นเมธอดพิเศษ เรียกว่า **constructor** — ถูกเรียกใช้**อัตโนมัติทันที**ตอนสร้าง object ใหม่ (ตอนเขียน `Student("มานี", 20, 3.75)`) มีหน้าที่ "ตั้งค่าเริ่มต้น" ให้ attribute ต่างๆ ของ object

### 2.2 `self` คืออะไร?

`self` หมายถึง **"object ตัวนี้เอง"** — ใช้อ้างอิงถึง object ที่กำลังถูกสร้าง/เรียกใช้งานอยู่ ทุกเมธอดใน class (รวมถึง `__init__`) ต้องมี `self` เป็น parameter ตัวแรกเสมอ (Python ใส่ให้อัตโนมัติตอนเรียกใช้ ไม่ต้องส่งเองตอน call)

```python

#                    ↑      ↑    ↑
#        self ไม่ต้องใส่ (Python ใส่ student1 ให้เอง)
#        name="มานี", age=20, gpa=3.75 ตรงกับ parameter ที่เหลือ
```

## 3. Attribute (คุณสมบัติ)

**Attribute** คือ "ข้อมูล" ที่เก็บอยู่ใน object — เทียบได้กับ "ตัวแปร" ที่ผูกอยู่กับ object นั้นโดยเฉพาะ เข้าถึงด้วย `object.attribute_name`

จากตัวอย่างข้างบน `name`, `age`, `gpa` คือ attribute ของ object `Student`


````python
class Student:
    def __init__(self, name, age, gpa, courses):
        self.name = name
        self.age = age
        self.gpa = gpa
        self.courses = courses   # attribute เป็น list ก็ได้ (เชื่อมกับ Collections ที่เรียนมา)

student1 = Student("มานี", 20, 3.75, ["SC612104", "SC612105", "SC612106"])

print(student1.name)
print(student1.courses)            # attribute ที่เป็น list
print(len(student1.courses))       # ใช้ฟังก์ชันที่เรียนมากับ attribute ได้ปกติ
print(student1.courses[0])         # indexing list ที่เป็น attribute
````

### 3.1 การแก้ไข attribute

แก้ไขค่า attribute ได้โดยตรงผ่าน `object.attribute_name = ค่าใหม่`


````python
class Student:
    def __init__(self, name, age, gpa):
        self.name = name
        self.age = age
        self.gpa = gpa

student1 = Student("มานี", 20, 3.75)
print("ก่อนแก้ไข:", student1.gpa)

student1.gpa = 3.95   # แก้ไข attribute โดยตรง
print("หลังแก้ไข:", student1.gpa)

student1.age += 1     # เพิ่มอายุ 1 ปี (เหมือนวันเกิด)
print("อายุใหม่:", student1.age)
````

## 4. Method (พฤติกรรม)

**Method** คือ "ฟังก์ชันที่อยู่ข้างใน class" — ทำงานกับ attribute ของ object นั้นได้โดยตรงผ่าน `self` เปรียบเทียบง่ายๆ:

- **Attribute** = สิ่งที่ object "มี" (เช่น ชื่อ, อายุ, GPA)
- **Method** = สิ่งที่ object "ทำได้" (เช่น แสดงข้อมูลตัวเอง, คำนวณบางอย่าง)

### โครงสร้าง method

```python
class ชื่อClass:
    def __init__(self, ...):
        ...

    def method_name(self, parameter1, ...):    # ต้องมี self เป็นตัวแรกเสมอ
        # ใช้ self.attribute เพื่อเข้าถึงข้อมูลของ object นี้
        return ผลลัพธ์
```


````python
class Student:
    def __init__(self, name, age, gpa):
        self.name = name
        self.age = age
        self.gpa = gpa

    def introduce(self):                          # method ไม่มี parameter เพิ่ม (มีแค่ self)
        print(f"สวัสดีครับ ผมชื่อ {self.name} อายุ {self.age} ปี")

    def is_honor_student(self):                    # method ที่ return ค่า
        return self.gpa >= 3.5

    def add_gpa_bonus(self, bonus):                 # method ที่รับ parameter เพิ่ม
        self.gpa = min(self.gpa + bonus, 4.0)        # บวกคะแนนพิเศษ แต่ไม่เกิน 4.0
        print(f"GPA ใหม่ของ {self.name}: {self.gpa}")

student1 = Student("มานี", 20, 3.75)

student1.introduce()                                 # เรียกใช้ method (ไม่ต้องส่ง self เอง)
print("เกียรตินิยมไหม:", student1.is_honor_student())
student1.add_gpa_bonus(0.5)
````

**สังเกตความแตกต่างจาก function ทั่วไป:** Method ต้องเรียกผ่าน `object.method_name()` เสมอ (ไม่เรียกแบบ `method_name()` ตรงๆ แบบ function ปกติ) และ Python จะส่ง `self` (คือ `student1`) ให้อัตโนมัติ — เราไม่ต้องใส่ argument แทน `self` เอง

### 4.1 ตัวอย่างที่สมบูรณ์: class ที่มี attribute เป็น list และ method ที่จัดการ list นั้น (เชื่อมกับ Collections)


````python
class Student:
    def __init__(self, name, gpa):
        self.name = name
        self.gpa = gpa
        self.courses = []     # attribute เริ่มต้นเป็น list ว่าง

    def enroll(self, course_name):                # method เพิ่มวิชาเข้า list
        self.courses.append(course_name)
        print(f"{self.name} ลงทะเบียน {course_name} แล้ว")

    def drop(self, course_name):                   # method ลบวิชาออกจาก list
        if course_name in self.courses:
            self.courses.remove(course_name)
            print(f"{self.name} ถอนวิชา {course_name} แล้ว")
        else:
            print(f"{self.name} ไม่ได้ลงวิชา {course_name} อยู่")

    def show_courses(self):                         # method แสดงรายวิชาทั้งหมด
        print(f"วิชาที่ {self.name} ลงทะเบียน: {self.courses}")
        print(f"จำนวนวิชาทั้งหมด: {len(self.courses)}")

student1 = Student("มานี", 3.75)
student1.enroll("SC612104")
student1.enroll("SC612105")
student1.enroll("SC612106")
student1.show_courses()

student1.drop("SC612105")
student1.show_courses()
````

## 5. ตัวอย่าง OOP ที่ใช้งานจริง: ระบบจัดการนักศึกษาทั้งห้อง

มาดูตัวอย่างที่รวมทุกอย่างเข้าด้วยกัน — class ที่มี attribute หลายตัว method หลายตัว และใช้ class ร่วมกับ list (เก็บหลาย object ไว้ใน list เดียว)


````python
class Student:
    def __init__(self, student_id, name, gpa):
        self.student_id = student_id
        self.name = name
        self.gpa = gpa

    def is_honor_student(self):
        return self.gpa >= 3.5

    def get_grade_level(self):
        if self.gpa >= 3.5:
            return "เกียรตินิยม"
        elif self.gpa >= 2.0:
            return "ปกติ"
        else:
            return "ต้องปรับปรุง"

    def __str__(self):
        # เมธอดพิเศษ: กำหนดว่าเวลา print(object) นี้ จะแสดงผลแบบไหน
        return f"[{self.student_id}] {self.name} (GPA: {self.gpa}, {self.get_grade_level()})"


# สร้าง list ของ object (เชื่อมกับ list ที่เรียนมาจาก Collections)
students = [
    Student("65010001", "มานี", 3.75),
    Student("65010002", "สมชาย", 3.20),
    Student("65010003", "วิชัย", 1.95),
    Student("65010004", "สมหญิง", 3.60),
]

# วน for loop ผ่าน list ของ object (เหมือนวน list ปกติ)
for student in students:
    print(student)   # เพราะมี __str__ จึง print object ได้ตรงๆ อ่านง่าย
````

**สังเกต:** `__str__` เป็นอีกหนึ่งเมธอดพิเศษ (เหมือน `__init__`) — กำหนดว่าเมื่อเอา object ไป `print()` ตรงๆ จะแสดงข้อความอะไร ถ้าไม่มี `__str__` Python จะแสดงแค่ `<__main__.Student object at 0x...>` ซึ่งอ่านไม่รู้เรื่อง

### 5.1 ใช้ method ของ object ร่วมกับ list comprehension และฟังก์ชันที่เรียนมาแล้ว


````python
# กรองหาเฉพาะนักศึกษาเกียรตินิยม (เชื่อมกับ list comprehension)
honor_students = [s for s in students if s.is_honor_student()]

print("รายชื่อนักศึกษาเกียรตินิยม:")
for s in honor_students:
    print(f"  - {s.name} (GPA {s.gpa})")

print()

# หา GPA เฉลี่ยของทั้งห้อง (เชื่อมกับ list comprehension + sum/len)
all_gpas = [s.gpa for s in students]
average_gpa = sum(all_gpas) / len(all_gpas)
print(f"GPA เฉลี่ยทั้งห้อง: {average_gpa:.2f}")

# หานักศึกษาที่ GPA สูงสุด (เชื่อมกับ max() และ key parameter)
top_student = max(students, key=lambda s: s.gpa)
print(f"นักศึกษา GPA สูงสุด: {top_student}")
````

### ข้อ 1: OOP — class BankAccount
สร้าง class `BankAccount` ที่มี:
1. `__init__(self, owner, balance=0)` — attribute `owner` และ `balance`
2. method `deposit(self, amount)` — เพิ่มเงินเข้า balance
3. method `withdraw(self, amount)` — ลดเงินออกจาก balance (ถ้า amount มากกว่า balance ให้ print ว่า "เงินไม่พอ" และไม่หักเงิน)
4. method `show_balance(self)` — แสดงยอดเงินปัจจุบัน

ทดสอบสร้าง object แล้วลองเรียก deposit/withdraw หลายครั้ง


````python
# เขียนคำตอบที่นี่

class Bankaccount:
    def __init__ (self, owner, balance=0):
        self.owner = owner
        self.balance = balance


    def deposit(self, amount):
       self.balance += amount
       print(f"เพิ่มเงินเข้า",amount)

    def withdraw(self, amount):
      if self.balance < amount:
          return print("not enough")

      else:
        self.balance -= amount
        print(f"withdraw",{amount},"now have",{self.balance})

    def show_balance(self):
      print(f"ยอดเงินปัจจุบัน",self.balance)



my_account = Bankaccount(f"kik",500)
my_account.deposit(700)
my_account.withdraw(400)
my_account.show_balance()





````

### ข้อ 2: OOP — เชื่อมกับ list ของ object
จาก class `BankAccount` ในข้อ 3 ให้สร้าง object 3 บัญชี เก็บไว้ใน list จากนั้น:
1. วน loop deposit เงิน 100 บาทให้ทุกบัญชี
2. ใช้ list comprehension หาบัญชีที่มี balance มากกว่า 500 บาท
3. หาบัญชีที่มี balance สูงสุด (Hint: ใช้ `max()` กับ parameter `key`)


````python
# เขียนคำตอบที่นี่ เหมือนข้อ5
class BankAccount:
    def __init__ (self, name, balance):
        self.name = name
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount

    def withdraw(self,balance):
      if amount <= self.balance:
        self.balance -= amount
      else:
        print("ยอดเงินไม่พอ")

    def __repr__(self):
        return f"{self.name}: {self.balance}"



accounts = [
    BankAccount("song", 250),
    BankAccount("sam", 870),
    BankAccount("si", 560)]

for acc in accounts:
    acc.deposit(100)

rich_accounts = [acc for acc in accounts if acc.balance > 500]

max_account = max(accounts, key=lambda acc: acc.balance)

print("หลังจากเพิ่มเงิน100ทุกบัญชีทั้งหมด", accounts)
print("บัญชีที่มีเงินมากกว่า 500", rich_accounts)
print("บัญชีที่มีเงินสูงสุด", max_account)
````

4. ใน list ของ animals ที่สร้างไว้ ให้วนเพื่อหาว่า สัตว์ทั้งหมด มีกี่ตัวที่มีขนาดต่าง ๆ

````python
#ใช้ for loop
class Animal:
  def __init__(self, name, color, size):
    self.name = name
    self.color = color
    self.size = size

  def makeSound():
    return "เสียงสัตว์"

  def __str__(self):
    return f"{self.name} ({self.color},{self.size})"

class Dog(Animal):
  def makeSound(self):
    return "โฮ่งงงง"

class Cat(Animal):
  def makeSound(self):
    return "เมี้ยวววว"

class Lion(Animal):
  def makeSound(self):
    return "กรรรรร"

animals = [
    Dog("miu","Orange","small"),
    Cat("somO","white","medium"),
    Lion("tanu","black","large")
]

for animal in animals:
    print(f"{animal} → {animal.makeSound()}")


size_count = {"small": 0, "medium": 0, "large": 0}
for animal in animals:
    size_count[animal.size] += 1


print("จำนวนสัตว์แต่ละขนาด:")
for size, count in size_count.items():
    print(f"{size}: {count}")
````
